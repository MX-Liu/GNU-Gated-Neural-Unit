```python
import os
import pandas as pd
import numpy as np

# --- Configuration ---
folder_name = 'logs_cluster'
dataset = 'cifar100'
ds_percentage = 1.0
noise = 0.0
# model_list = ['resnet18', 'resnet34', 'resnet50', 'vgg16', 'densenet121', 'densenet169', 'densenet201', 'densenet161', 'mobilenet_v2', 'googlenet']
model_list = ['resnet18', 'resnet34', 'vgg16', 'densenet121', 'densenet169', 'densenet201', 'mobilenet_v2', 'googlenet']
base_classifier_list = ['mlp', 'kan', 'gnu']
degree = 3
grid = 10
alpha1 = 0.5
alpha2 = 0.5
num_epochs = 100
seeds = [0, 1, 2]

# Define column width for alignment.
COL_WIDTH = 18

# --- Helper Functions ---

def bold(text):
    """Wraps text in ANSI escape codes for bold printing in a terminal."""
    return f"\033[1m{text}\033[0m"

def read_results(log_file):
    """Reads the results from the log file and returns a dictionary of metrics."""
    if not os.path.exists(log_file):
        return None
    try:
        with open(log_file, 'r') as file:
            lines = file.readlines()
        if not lines: return None
        last_line = lines[-1].strip()
        metrics = last_line.split(', ')
        results = {key: float(value.replace('%', '')) for metric in metrics for key, value in [metric.split(': ')]}
        return results
    except Exception as e:
        print(f"Warning: Could not parse file {log_file}. Error: {e}")
        return None

def get_results_for_config(classifiers_to_run, file_pattern):
    """
    Collects results for a single configuration and returns mean and formatted DataFrames.
    """
    data = []
    for classifier in classifiers_to_run:
        row = {'classifier': classifier}
        for model in model_list:
            accuracies = [
                res['accuracy'] for seed in seeds 
                if (res := read_results(file_pattern.format(model=model, classifier=classifier, seed=seed))) and 'accuracy' in res
            ]
            if accuracies:
                row[f"{model}_mean"] = np.mean(accuracies)
                row[f"{model}_std"] = np.std(accuracies)
        data.append(row)

    if not data:
        return pd.DataFrame(), pd.DataFrame()

    df = pd.DataFrame(data).set_index('classifier')
    
    # Create the two required DataFrames
    mean_df = df[[f"{m}_mean" for m in model_list]].copy()
    mean_df.columns = model_list # Rename columns to just model names

    formatted_df = pd.DataFrame(index=df.index)
    for model in model_list:
        mean_col = f"{model}_mean"
        std_col = f"{model}_std"
        if mean_col in df.columns:
            formatted_df[model] = df.apply(
                lambda r: f"{r[mean_col]:.2f} ± {r[std_col]:.2f}" if pd.notna(r[mean_col]) else "N/A",
                axis=1
            )
        else:
            formatted_df[model] = "N/A"
            
    return mean_df, formatted_df

def print_final_tables(all_results, global_best_locations):
    """
    Prints all tables, using the global_best_locations to bold the best cell per model.
    """
    for title, (mean_df, formatted_df) in all_results.items():
        print(f"\n--- {title} ---")
        if formatted_df.empty:
            print("No results to display.")
            continue

        # Print Headers
        headers = ["Classifier"] + model_list
        header_str = "".join([f"{h:<{COL_WIDTH}}" for h in headers])
        print(header_str)
        print("-" * len(header_str))

        # Print Data Rows
        for classifier, row in formatted_df.iterrows():
            row_str = f"{classifier:<{COL_WIDTH}}"
            for model in model_list:
                cell_str = row[model]
                
                # Check if this cell is the global maximum for its model column
                best_location_for_model = global_best_locations.get(model)
                if best_location_for_model and best_location_for_model == (title, classifier):
                    # Pad the string *before* adding bold codes to maintain alignment
                    padded_cell = f"{cell_str:<{COL_WIDTH}}"
                    row_str += bold(padded_cell)
                else:
                    row_str += f"{cell_str:<{COL_WIDTH}}"
            print(row_str)


# --- Main Execution ---

if __name__ == "__main__":
    # 1. Define all configurations to run
    configs = [
        {
            "title": "With Linear Projection (activation: silu)",
            "classifiers": base_classifier_list,
            "pattern": f"{folder_name}/Dataset_{dataset}_percentage_{ds_percentage}_noise_{noise}_"
                       f"model_{{model}}_classifier_{{classifier}}_activation_silu_"
                       f"degree_{degree}_grid_{grid}_alp1_{alpha1}_alp2_{alpha2}_"
                       f"epoch_{num_epochs}_projection_seed_{{seed}}.log"
        },
        {
            "title": "With Linear Projection (activation: identity)",
            "classifiers": base_classifier_list,
            "pattern": f"{folder_name}/Dataset_{dataset}_percentage_{ds_percentage}_noise_{noise}_"
                       f"model_{{model}}_classifier_{{classifier}}_activation_identity_"
                       f"degree_{degree}_grid_{grid}_alp1_{alpha1}_alp2_{alpha2}_"
                       f"epoch_{num_epochs}_projection_seed_{{seed}}.log"
        },
        {
            "title": "Without Linear Projection (activation: silu)",
            "classifiers": ['gnu'],
            "pattern": f"{folder_name}/Dataset_{dataset}_percentage_{ds_percentage}_noise_{noise}_"
                       f"model_{{model}}_classifier_{{classifier}}_activation_silu_"
                       f"degree_{degree}_grid_{grid}_alp1_{alpha1}_alp2_{alpha2}_"
                       f"epoch_{num_epochs}_seed_{{seed}}.log"
        },
        {
            "title": "Without Linear Projection (activation: identity)",
            "classifiers": ['gnu'],
            "pattern": f"{folder_name}/Dataset_{dataset}_percentage_{ds_percentage}_noise_{noise}_"
                       f"model_{{model}}_classifier_{{classifier}}_activation_identity_"
                       f"degree_{degree}_grid_{grid}_alp1_{alpha1}_alp2_{alpha2}_"
                       f"epoch_{num_epochs}_seed_{{seed}}.log"
        }
    ]

    # 2. Collect data from all configurations
    all_results = {}
    all_mean_dfs = []
    for config in configs:
        mean_df, formatted_df = get_results_for_config(config['classifiers'], config['pattern'])
        if not mean_df.empty:
            all_results[config['title']] = (mean_df, formatted_df)
            
            # Prepare df for global comparison by adding the source title as a column
            temp_df = mean_df.copy()
            temp_df['source_title'] = config['title']
            all_mean_dfs.append(temp_df)

    # 3. Find the global best location for each model
    global_best_locations = {}
    if all_mean_dfs:
        # Combine all mean DataFrames into one
        combined_df = pd.concat(all_mean_dfs)
        # The index is the classifier name, so reset it to be a column
        combined_df.reset_index(inplace=True)
        combined_df = combined_df.rename(columns={'index': 'classifier'})
        
        # Now find the index of the max value for each model column
        for model in model_list:
            if model in combined_df.columns and combined_df[model].notna().any():
                # idxmax() gives the integer index of the row with the max value
                best_idx = combined_df[model].idxmax()
                best_row = combined_df.loc[best_idx]
                # Store the location as a tuple: (table_title, classifier)
                global_best_locations[model] = (best_row['source_title'], best_row['classifier'])

    # 4. Print all tables using the calculated global best locations
    print_final_tables(all_results, global_best_locations)
```

```python
import os
import pandas as pd
import numpy as np

# --- Configuration ---
folder_name = 'logs_cluster'
dataset = 'cifar100'
ds_percentage = 0.1
noise = 0.0
# model_list = ['resnet18', 'resnet34', 'resnet50', 'vgg16', 'densenet121', 'densenet169', 'densenet201', 'densenet161', 'mobilenet_v2', 'googlenet']
model_list = ['resnet18', 'resnet34', 'vgg16', 'densenet121', 'densenet169', 'densenet201', 'mobilenet_v2', 'googlenet']
base_classifier_list = ['mlp', 'kan', 'gnu']
degree = 3
grid = 10
alpha1 = 0.5
alpha2 = 0.5
num_epochs = 100
seeds = [0, 1, 2]

# Define column width for alignment.
COL_WIDTH = 18

# --- Helper Functions ---

def bold(text):
    """Wraps text in ANSI escape codes for bold printing in a terminal."""
    return f"\033[1m{text}\033[0m"

def read_results(log_file):
    """Reads the results from the log file and returns a dictionary of metrics."""

    if not os.path.exists(log_file):
        log_file = log_file.replace("0.5", "1.0")

    if not os.path.exists(log_file):
        return None
    try:
        with open(log_file, 'r') as file:
            lines = file.readlines()
        if not lines: return None
        last_line = lines[-1].strip()
        metrics = last_line.split(', ')
        results = {key: float(value.replace('%', '')) for metric in metrics for key, value in [metric.split(': ')]}

        return results
    except Exception as e:
        print(f"Warning: Could not parse file {log_file}. Error: {e}")
        return None

def get_results_for_config(classifiers_to_run, file_pattern):
    """
    Collects results for a single configuration and returns mean and formatted DataFrames.
    """
    data = []
    for classifier in classifiers_to_run:
        row = {'classifier': classifier}
        for model in model_list:
            accuracies = [
                res['accuracy'] for seed in seeds
                if (res := read_results(file_pattern.format(model=model, classifier=classifier, seed=seed))) and 'accuracy' in res
            ]
            if accuracies:
                row[f"{model}_mean"] = np.mean(accuracies)
                row[f"{model}_std"] = np.std(accuracies)
        data.append(row)

    if not data:
        return pd.DataFrame(), pd.DataFrame()

    df = pd.DataFrame(data).set_index('classifier')

    # Create the two required DataFrames
    mean_df = df[[f"{m}_mean" for m in model_list]].copy()
    mean_df.columns = model_list # Rename columns to just model names

    formatted_df = pd.DataFrame(index=df.index)
    for model in model_list:
        mean_col = f"{model}_mean"
        std_col = f"{model}_std"
        if mean_col in df.columns:
            formatted_df[model] = df.apply(
                lambda r: f"{r[mean_col]:.2f} ± {r[std_col]:.2f}" if pd.notna(r[mean_col]) else "N/A",
                axis=1
            )
        else:
            formatted_df[model] = "N/A"

    return mean_df, formatted_df

def print_final_tables(all_results, global_best_locations):
    """
    Prints all tables, using the global_best_locations to bold the best cell per model.
    """
    for title, (mean_df, formatted_df) in all_results.items():
        print(f"\n--- {title} ---")
        if formatted_df.empty:
            print("No results to display.")
            continue

        # Print Headers
        headers = ["Classifier"] + model_list
        header_str = "".join([f"{h:<{COL_WIDTH}}" for h in headers])
        print(header_str)
        print("-" * len(header_str))

        # Print Data Rows
        for classifier, row in formatted_df.iterrows():
            row_str = f"{classifier:<{COL_WIDTH}}"
            for model in model_list:
                cell_str = row[model]

                # Check if this cell is the global maximum for its model column
                best_location_for_model = global_best_locations.get(model)
                if best_location_for_model and best_location_for_model == (title, classifier):
                    # Pad the string *before* adding bold codes to maintain alignment
                    padded_cell = f"{cell_str:<{COL_WIDTH}}"
                    row_str += bold(padded_cell)
                else:
                    row_str += f"{cell_str:<{COL_WIDTH}}"
            print(row_str)


# --- Main Execution ---

if __name__ == "__main__":
    # 1. Define all configurations to run
    configs = [
        {
            "title": "With Linear Projection (activation: silu)",
            "classifiers": ['mlp', 'kan'],  # Removed 'gnu' from this list
            "pattern": f"{folder_name}/Dataset_{dataset}_percentage_{ds_percentage}_noise_{noise}_"
                       f"model_{{model}}_classifier_{{classifier}}_activation_silu_"
                       f"degree_{degree}_grid_{grid}_alp1_{alpha1}_alp2_{alpha2}_"
                       f"epoch_{num_epochs}_seed_{{seed}}.log"
        },
        # {
        #     "title": "With Linear Projection (activation: identity)",
        #     "classifiers": ['mlp', 'kan'], # Removed 'gnu' from this list as well for consistency
        #     "pattern": f"{folder_name}/Dataset_{dataset}_percentage_{ds_percentage}_noise_{noise}_"
        #                f"model_{{model}}_classifier_{{classifier}}_activation_identity_"
        #                f"degree_{degree}_grid_{grid}_alp1_{alpha1}_alp2_{alpha2}_"
        #                f"epoch_{num_epochs}_projection_seed_{{seed}}.log"
        # },
        {
            "title": "Without Linear Projection (activation: silu)",
            "classifiers": ['gnu'],
            "pattern": f"{folder_name}/Dataset_{dataset}_percentage_{ds_percentage}_noise_{noise}_"
                       f"model_{{model}}_classifier_{{classifier}}_activation_silu_"
                       f"degree_{degree}_grid_{grid}_alp1_{alpha1}_alp2_{alpha2}_"
                       f"epoch_{num_epochs}_seed_{{seed}}.log"
        },
        {
            "title": "Without Linear Projection (activation: identity)",
            "classifiers": ['gnu'],
            "pattern": f"{folder_name}/Dataset_{dataset}_percentage_{ds_percentage}_noise_{noise}_"
                       f"model_{{model}}_classifier_{{classifier}}_activation_identity_"
                       f"degree_{degree}_grid_{grid}_alp1_{alpha1}_alp2_{alpha2}_"
                       f"epoch_{num_epochs}_seed_{{seed}}.log"
        }
    ]

    # 2. Collect data from all configurations
    all_results = {}
    all_mean_dfs = []
    for config in configs:
        mean_df, formatted_df = get_results_for_config(config['classifiers'], config['pattern'])
        if not mean_df.empty:
            all_results[config['title']] = (mean_df, formatted_df)

            # Prepare df for global comparison by adding the source title as a column
            temp_df = mean_df.copy()
            temp_df['source_title'] = config['title']
            all_mean_dfs.append(temp_df)

    # 3. Find the global best location for each model
    global_best_locations = {}
    if all_mean_dfs:
        # Combine all mean DataFrames into one
        combined_df = pd.concat(all_mean_dfs)
        # The index is the classifier name, so reset it to be a column
        combined_df.reset_index(inplace=True)
        combined_df = combined_df.rename(columns={'index': 'classifier'})

        # Now find the index of the max value for each model column
        for model in model_list:
            if model in combined_df.columns and combined_df[model].notna().any():
                # idxmax() gives the integer index of the row with the max value
                best_idx = combined_df[model].idxmax()
                best_row = combined_df.loc[best_idx]
                # Store the location as a tuple: (table_title, classifier)
                global_best_locations[model] = (best_row['source_title'], best_row['classifier'])

    # 4. Print all tables using the calculated global best locations
    print_final_tables(all_results, global_best_locations)
```

```python
import os
import pandas as pd
import numpy as np

# --- Configuration ---
folder_name = 'logs_cluster'
dataset = 'cifar100'
ds_percentage = 0.1
noise = 0.0
model_list = ['resnet18', 'resnet34', 'vgg16', 'densenet121', 'densenet169', 'densenet201', 'mobilenet_v2', 'googlenet']
base_classifier_list = ['mlp', 'kan', 'gnu']
degree = 3
grid = 10
alpha1 = 0.5
alpha2 = 0.5
num_epochs = 100
seeds = [0, 1, 2]

# --- Helper Functions ---

def read_results(log_file):
    """Reads the results from the log file and returns a dictionary of metrics."""
    if not os.path.exists(log_file):
        log_file = log_file.replace("0.5", "1.0")

    if not os.path.exists(log_file):
        return None
    try:
        with open(log_file, 'r') as file:
            lines = file.readlines()
        if not lines: return None
        last_line = lines[-1].strip()
        metrics = last_line.split(', ')
        results = {key: float(value.replace('%', '')) for metric in metrics for key, value in [metric.split(': ')]}
        return results
    except Exception as e:
        # Using print to stderr for warnings is a good practice
        import sys
        print(f"Warning: Could not parse file {log_file}. Error: {e}", file=sys.stderr)
        return None

def get_results_for_config(classifiers_to_run, file_pattern):
    """
    Collects results for a single configuration and returns mean and formatted DataFrames.
    """
    data = []
    for classifier in classifiers_to_run:
        row = {'classifier': classifier}
        for model in model_list:
            accuracies = [
                res['accuracy'] for seed in seeds
                if (res := read_results(file_pattern.format(model=model, classifier=classifier, seed=seed))) and 'accuracy' in res
            ]
            if accuracies:
                row[f"{model}_mean"] = np.mean(accuracies)
                row[f"{model}_std"] = np.std(accuracies)
        data.append(row)

    if not data:
        return pd.DataFrame(), pd.DataFrame()

    df = pd.DataFrame(data).set_index('classifier')
    
    # Create the mean DataFrame for calculations
    mean_df_cols = [f"{m}_mean" for m in model_list if f"{m}_mean" in df.columns]
    mean_df = df[mean_df_cols].copy()
    mean_df.columns = [col.replace('_mean', '') for col in mean_df.columns]

    # Create the formatted DataFrame for display
    formatted_df = pd.DataFrame(index=df.index)
    for model in model_list:
        mean_col = f"{model}_mean"
        std_col = f"{model}_std"
        if mean_col in df.columns and std_col in df.columns:
            formatted_df[model] = df.apply(
                lambda r: f"{r[mean_col]:.2f}({r[std_col]:.2f})" if pd.notna(r[mean_col]) else "N/A",
                axis=1
            )
        else:
            formatted_df[model] = "N/A"
            
    return mean_df, formatted_df

def generate_latex_tables(all_results, global_best_locations):
    """
    Generates a single string containing all tables in LaTeX format with mean(std) format.
    """
    latex_code_list = [
        "% Add these packages to your LaTeX preamble:",
        "% \\usepackage{booktabs}",
        "% \\usepackage{graphicx}",
        "% \\usepackage{amsmath}",
    ]

    for title, (mean_df, formatted_df) in all_results.items():
        if formatted_df.empty:
            continue

        table_label = "tab:" + title.lower().replace(" ", "-").replace(":", "").replace("(", "").replace(")", "")
        latex_headers = [h.replace('_', r'\_') for h in model_list]
        column_format = 'l' + 'c' * len(model_list)
        
        table_lines = [
            f"\\begin{{table*}}[t!]",
            f"\\centering",
            f"\\caption{{{title}}}",
            f"\\label{{{table_label}}}",
            f"\\resizebox{{\\textwidth}}{{!}}{{%",
            f"\\begin{{tabular}}{{{column_format}}}",
            f"\\toprule",
            "\\textbf{Classifier} & " + " & ".join(f"\\textbf{{{h}}}" for h in latex_headers) + " \\\\",
            f"\\midrule"
        ]

        for classifier, row in formatted_df.iterrows():
            safe_classifier = classifier.replace('_', r'\_')
            row_data = [safe_classifier]
            for model in model_list:
                cell_str = row[model]
                
                best_location = global_best_locations.get(model)
                if best_location and best_location == (title, classifier):
                    cell_str = f"\\textbf{{{cell_str}}}"
                
                row_data.append(cell_str)
            
            table_lines.append(" & ".join(row_data) + " \\\\")

        table_lines.extend([
            f"\\bottomrule",
            f"\\end{{tabular}}%",
            f"}}",
            f"\\end{{table*}}"
        ])
        
        latex_code_list.append("\n".join(table_lines))
        
    return "\n\n".join(latex_code_list)


# --- Main Execution ---

if __name__ == "__main__":
    # 1. Define all configurations to run
    configs = [
        {
            "title": "With Linear Projection (activation: silu)",
            "classifiers": ['mlp', 'kan'],
            "pattern": f"{folder_name}/Dataset_{dataset}_percentage_{ds_percentage}_noise_{noise}_"
                       f"model_{{model}}_classifier_{{classifier}}_activation_silu_"
                       f"degree_{degree}_grid_{grid}_alp1_{alpha1}_alp2_{alpha2}_"
                       f"epoch_{num_epochs}_seed_{{seed}}.log"
        },
        {
            "title": "Without Linear Projection (activation: silu)",
            "classifiers": ['gnu'],
            "pattern": f"{folder_name}/Dataset_{dataset}_percentage_{ds_percentage}_noise_{noise}_"
                       f"model_{{model}}_classifier_{{classifier}}_activation_silu_"
                       f"degree_{degree}_grid_{grid}_alp1_{alpha1}_alp2_{alpha2}_"
                       f"epoch_{num_epochs}_seed_{{seed}}.log"
        },
        {
            "title": "Without Linear Projection (activation: identity)",
            "classifiers": ['gnu'],
            "pattern": f"{folder_name}/Dataset_{dataset}_percentage_{ds_percentage}_noise_{noise}_"
                       f"model_{{model}}_classifier_{{classifier}}_activation_identity_"
                       f"degree_{degree}_grid_{grid}_alp1_{alpha1}_alp2_{alpha2}_"
                       f"epoch_{num_epochs}_seed_{{seed}}.log"
        }
    ]

    # 2. Collect data from all configurations
    all_results = {}
    all_mean_dfs = []
    for config in configs:
        mean_df, formatted_df = get_results_for_config(config['classifiers'], config['pattern'])
        if not mean_df.empty:
            all_results[config['title']] = (mean_df, formatted_df)
            temp_df = mean_df.copy()
            temp_df['source_title'] = config['title']
            all_mean_dfs.append(temp_df)

    # 3. Find the global best location for each model
    global_best_locations = {}
    if all_mean_dfs:
        combined_df = pd.concat(all_mean_dfs).reset_index().rename(columns={'index': 'classifier'})
        for model in model_list:
            if model in combined_df.columns and combined_df[model].notna().any():
                best_idx = combined_df[model].idxmax()
                best_row = combined_df.loc[best_idx]
                global_best_locations[model] = (best_row['source_title'], best_row['classifier'])

    # 4. Generate and print the LaTeX tables
    latex_output = generate_latex_tables(all_results, global_best_locations)
    print(latex_output)
```
